```python
import pandas as pd
import numpy as np

# Method to determine interval estimate
def interval_estimate(sample_mean,z_alpha,size,sd):
    ie1 = sample_mean - (z_alpha*(sd/(size)**0.5))
    ie2 = sample_mean + (z_alpha*(sd/(size)**0.5))
    ans = f'Interval Estimate is [{ie1},{ie2}]'
    return ans

# Method to test hypothesis
def hypo_test(sample_mean,population_mean,z_alpha,sd,size):
    z_stats = (sample_mean - population_mean)/(sd/(size)**0.5)
    if(size>30):
        if(z_stats>z_alpha):
            ans = 'This is Z distribution and reject Null Hypothesis'
        else:
            ans = 'This is Z distribution and accept Null Hypothesis'
        return ans
    else:
        if(z_stats>z_alpha):
            ans = 'This is T distribution and reject Null Hypothesis'
        else:
            ans = 'This is T distribution and accept Null Hypothesis'
        return ans    

# stats_calc method 
def stats_calc(population_mean,alpha,sample_mean=0,sd=0,size=0,csv=None):
    s="""
    Enter 1 for Point Estimate
    Enter 2 for Interval Estimate
    Enter 3 for Hypothesis Testing
    Enter 4 for All
    """
    try:
        inp = int(input(s))
    except:
        print('Invalid Input')
        return
    
# Defined 2 dictionaries for z and t test
    z_dict = {0.05:1.68,0.025:1.96,0.010:2.32,0.005:2.57}
    t_dict = {15:{0.05:1.753,0.025:2.131,0.010:2.602,0.005:2.947},40:{0.05:1.684,0.025:2.02,0.010:2.42,0.005:2.70}}

# Condition where csv file is provided by the user
    if(csv):
        data = pd.read_csv(csv)
        size = len(data)
        if((size!=16) & (size!=41)):
            print('Sample size can be 16 or 41')
            return
        sample_mean = np.mean(data)[0]
        sd = np.std(data)[0]
        
# Calling functions according to the input given by the user
    if(inp==1):
        print('-'*100)
        print(f'Point Estimate is {sample_mean}')
    elif(inp==2):
        try:
            tail = int(input("Enter 1 for one-tail test and 2 for two-tail test"))
        except:
            print('Invalid tail value')

# Performing Z test if sample size>30
        if(size>30):
            if(tail==1):
# Getting z_alpha value as per alpha value given by user from above defined dictionary
                z_alpha = z_dict[alpha]
            elif(tail==2):
                alpha=alpha/2
                z_alpha = z_dict[alpha]
            else:
                print('Tail value can be 1 or 2')
                return
            print(interval_estimate(sample_mean,z_alpha,size,sd))
            
# Performing T test for sample size <= 30
        else:
            if(tail==1):
# Getting t_alpha value as per alpha value and number of samples given by user from above defined dictionary
                t_alpha = t_dict[size-1][alpha]
            elif(tail==2):
                alpha=alpha/2
                t_alpha = t_dict[size-1][alpha]
            else:
                print('Tail value can be 1 or 2')
                return
            print('-'*100)
            print(interval_estimate(sample_mean,t_alpha,size,sd))
    elif(inp==3):
        try:
            tail = int(input("Enter 1 for one-tail test and 2 for two-tail test"))
        except:
            print('Invalid tail value')
        if(size>30):
            if(tail==1):
                alpha=0.05
                z_alpha = z_dict[alpha]
            elif(tail==2):
                alpha=0.025
                z_alpha = z_dict[alpha]
            else:
                print('Enter valid input for tail')
                return
            print('-'*100)
            print(hypo_test(sample_mean,population_mean,z_alpha,sd,size))
        else:
            if(tail==1):
                alpha=0.05
                t_alpha = t_dict[size-1][alpha]
            elif(tail==2):
                alpha=0.025
                t_alpha = t_dict[size-1][alpha]
            else:
                print('Enter valid input for tail')
                return
            print('-'*100)
            print(hypo_test(sample_mean,population_mean,t_alpha,sd,size))
    elif(inp==4):
        try:
            tail = int(input("Enter 1 for one-tail test and 2 for two-tail test"))
        except:
            print('Invalid tail value')
            return
        if(size>30):
            if(tail==1):
                z_alpha = z_dict[alpha]
            elif(tail==2):
                alpha=alpha/2
                z_alpha = z_dict[alpha]
            else:
                print('Tail value can be 1 or 2')
                return
            print('-'*100)
            print(f'Point Estimate is {sample_mean}')
            print(interval_estimate(sample_mean,z_alpha,size,sd))
            print(hypo_test(sample_mean,population_mean,z_alpha,sd,size))
        else:
            if(tail==1):
                t_alpha = t_dict[size-1][alpha]
            elif(tail==2):
                alpha=alpha/2
                t_alpha = t_dict[size-1][alpha]
            else:
                print('Tail value can be 1 or 2')
                return
            print('-'*100)
            print(f'Point Estimate is {sample_mean}')
            print(interval_estimate(sample_mean,t_alpha,size,sd))
            print(hypo_test(sample_mean,population_mean,t_alpha,sd,size))
        
        
    
```
