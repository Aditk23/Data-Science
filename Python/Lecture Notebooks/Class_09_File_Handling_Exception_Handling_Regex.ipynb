```python
import re
def validate_email(email):
    if '@' in email and '.' in email:
        print('Valid email id')
    else:
        print('Invalid Email id')
def password_valid(pwd): 
    if (re.search('[A-Z]', pwd) and re.search('[a-z]', pwd) and re.search('[0-9]', pwd) and re.search('[!@#$%^&*?]', pwd)):
        print('Valid Password')
    else:
        print('Invalid Password')  
def main():
    validate_email('Abc@gmail.com')
    password_valid('Abc@123')
    print(__name__)
        
if __name__ == '__main__':
    main()
```

# Exception Handling

```python
# try, except, else, finally, raise

# type of Errors
    1. Compile time error - Syantx Error, Indentation Error
    2. Run time error - NameError, ValueError, TypeError, FileNotFoundError etc
    
# Implicit
```

```python
Those Run time errors which can be handled using Exception Handling they are called Exceptions
```

```python
# Exceptions
try - The code which we expect to raise the error
except - If error is raised in try block then except block will be executed
else - if error is not raised then else block will be executed. (optional)
finally - default block which is executed irrespective of the error. We write the clean up code. (optional)
```

```python
num1 = int(input('Enter first number: '))
num2 = int(input('Enter second number: '))
try:
    res = num1/num2
except:
    print('Number is not divisible by 0')
else:
    print(res)
finally:
    print(num1)
    print(num2)
```
