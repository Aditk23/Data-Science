```python
data['tutions'] = data['tutions'].replace('none', 'No Tutions Taken')
```

```python
data = data.drop('index_to_dropped', 1)
```

```python
data.head(10)
```

```python
#Let's Explore the Gender Columns
list(data['gender'].unique())
```

```python
#Pie Chart

labels = list(data['gender'].unique())
size = list(data['gender'].value_counts())
fig1 = plt.figure(figsize=(20,10))

explode = (0.05, 0.05)

fig1, ax1 = plt.subplots()
ax1.pie(size, explode=explode, labels= labels, autopct = '%1.1f%%', shadow = False, startangle=90)
plt.show()
```

```python
#Let's check Section Column
data['section'].value_counts()
```

```python
list(data['section'].unique())
```

```python
#Bar Chart

x = list(data['section'].unique())
y = list(data['section'].value_counts())


x_pos = [i for i, _ in enumerate(x)]

plt.xlabel('Sections')
plt.ylabel('No. of Students')

plt.bar(x_pos, y, color='green')
plt.xticks(x_pos, x)
plt.show()

print(x_pos)
```

```python
plt.bar(x, y, color='green')
```

```python
#Let's explore Parental Level of Education
data['parental level of education'].value_counts(normalize=True)
```

```python
#Let's check Weight column
data["weight (in kgs)"].value_counts(bins = 5)
```

```python
#Violin Plot
ax = sns.violinplot(x=data["weight (in kgs)"])
```

```python
ax = sns.violinplot(x= 'gender', y = "weight (in kgs)", data = data)
```

```python
#Box Plots
ax = data.boxplot(column=['math score', 'english score', 'science score'])
```

```python
#Categorical Box Plot
sns.catplot(x='house', y='english score', data = data)
```

```python
#Statistics Functions
data['math score'].describe()
```

```python
#Check the Distribution of Data - Histograms

data['math score'].hist(bins=10, histtype = 'bar', align = 'mid', orientation = 'vertical', color = None)
```

```python
#Swarm Plot
ax = sns.swarmplot(x='section', y= 'english score', hue ='gender', data = data, dodge=True)
```

```python
#Unstack
data.groupby(['gender','section', 'house']).size().unstack()
```

```python
#Scatter Plot
plt.scatter(x = data['age'], y = data["height (in cms)"], color = 'blue', marker = 'v')
plt.show()
```

```python
#Looping to Visualize all Numerical Columns
for i in data.columns[6:12]:
    print(i)
    l = data.loc[:,i]
    sns.distplot(l, kde = True)
    plt.show()
```

```python
#Looping to Visualize all Numerical Columns
for i in data.columns[6:12]:
    print(i)
    l = data.loc[:,i]
    sns.boxplot(l)
    plt.show()
```

```python
#Point Plot
ax = sns.pointplot(x = 'section', y = 'english score', hue = 'gender', data = data)
```

```python
#Point Plot
ax = sns.pointplot(x = 'section', y = 'math score', hue = 'gender', data = data)
```

```python
#Heat Map for Correlation
sns.heatmap(data.corr(), fmt='.2g', annot=True, cmap = 'YlGnBu')
```

```python
#Regression Plot
sns.regplot(x='math score', y='english score', data = data, color = 'b', marker='o')
```
