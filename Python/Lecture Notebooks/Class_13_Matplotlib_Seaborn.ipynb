```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arr1 = np.linspace(1, 10, 20)
arr1
```

```python
arr2 = np.log(arr1)
arr2
```

```python
plt.plot(arr1, arr2, 'r')
plt.xlabel('Arr1')
plt.ylabel('Arr2')
plt.title('Arr1 vs Arr2')
```

```python
# Creating multiple plots
plt.subplot(1, 3, 1)
plt.plot(arr1, arr2, 'r--')
plt.xlabel('Arr1')
plt.ylabel('Arr2')

plt.subplot(1, 3, 2)
plt.plot(arr2, arr1, 'g:')

plt.subplot(1, 3, 2)
plt.plot(arr1, arr2, 'g:')
```

### Object Oriented Approach

```python
# Creating a figure object
fig = plt.figure()

# Adding axes to graph
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])                  # Left, bottom, width, height

ax.plot(arr1, arr2, 'r')
ax.set_xlabel('Arr1')
ax.set_ylabel('Arr2')
ax.set_title('Arr1 vs Arr2')
```

```python
# Creating plot inside another plot
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax2 = fig.add_axes([0.2, 0.5, 0.3, 0.4])

ax1.plot(arr2, arr1, 'r:')
ax2.plot(arr1, arr2, 'r:')
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax[0].plot(arr1, arr2, 'c-.')
ax[1].plot(arr1, arr2, 'm-')
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
ax[0][0].plot(arr1, 'r:')
ax[0][1].plot(arr2, 'g-')
ax[1][0].plot(arr1, arr2, 'y-.')
ax[1][1].plot(arr2, arr1, 'm--')

plt.tight_layout()
```

```python
# Figsize and DPI
fig, ax = plt.subplots(figsize = (10, 2),nrows = 2, ncols = 2, dpi = 100)
ax[0][0].plot(arr1, 'r:')
ax[0][1].plot(arr2, 'g-')
ax[1][0].plot(arr1, arr2, 'y-.')
ax[1][1].plot(arr2, arr1, 'm--')

plt.tight_layout()
```

```python
fig.savefig('plot.png')
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1,1])
ax.plot(arr1, arr2, 'r:', label = 'Arr1 vs Arr2')
ax.plot(arr2, arr1, 'g-.', label = 'Arr2 vs Arr1')
ax.legend(loc = 10)

```

```python
# color, ls, lw, marker, markersize, markerwidth, markerfacecolor, markeredgecolor

fig = plt.figure(figsize = (8, 5))

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(arr1, color = 'm', ls = ':', lw = 1)
ax.plot(arr1+2, color = 'r', ls = '--', lw = 2)
ax.plot(arr1+3, color = 'r', ls = '--', lw = 10)

```

```python
fig = plt.figure(figsize = (8, 5))

ax = fig.add_axes([0, 0, 1, 1])


ax.plot(arr1, color = 'r', ls = '-', lw = 2, markevery = 5, marker = 'o')
# ax.annotate()

ax.set_xticks([0, 4, 8, 12, 16,20])
ax.set_yticks([2.5, 5, 7.5, 10, 12.5])
```
